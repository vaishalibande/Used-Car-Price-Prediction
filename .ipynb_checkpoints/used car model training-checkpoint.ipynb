### Data Cleaning and Feature Engineering

```python
df_copy = df.copy().drop_duplicates()
df_copy
```

```python
# first strip the $ sign and then replace all occurances of a comma
# then convert the value to a integer
```

```python
cleaned_price = [int(str(value).strip("$").replace(',','')) for value in df_copy['price'].values]
```

```python
cleaned_price
```

```python
#now clean the miles
```

```python
miles = [int(str(m).replace('miles','').replace(',','').strip()) for m in df_copy['miles'].values]
```

```python
miles
```

```python
len(miles)
```

```python
# now name, color and condition : as these require splitting them into new coloumns:
```

```python
temp_name = df_copy['name'].copy()
temp_name = [str(name).split(" ", maxsplit=1) for name in temp_name.values]
make = [arr[0] for arr in temp_name]
len(make)
```

```python
temp_name
```

```python
make
```

```python
model = [arr[1] for arr in temp_name]
model
```

```python
len(model)
```

### color

```python
colors = [str(color).split(",") for color in df_copy['color'].values]
colors
```

```python
len(colors)
```

```python
exterior_color = [ str(color[0]).replace("exterior", "").strip() for color in colors ]
```

```python
len(exterior_color)
```

```python
interior_color = [str(color[1]).replace('interior','').strip() for color in colors]
len(interior_color)
```

```python
# now, let's deal with the condition variable that tells us important information about the number of accidents and the amount of owners
```

```python
conditions = [str(condition).split(",") for condition in df_copy['condition'].values]
conditions
```

```python
# now lets make accident in int
```

```python
accidents = [int(str(condition[0]).replace('No','0').replace('accident reported','').replace('accidents reported','').strip())
             for condition in conditions]
len(accidents)
```

```python
# owners = [int(str(condition[1]).replace('Owner','').replace('Owners','').strip()) for condition in conditions]
owners = [ int(str(condition[1]).replace("Owners", "").replace("Owner", "").strip()) for condition in conditions]
```

```python
len(owners)
```

```python
print(
    len(make), 
    len(model),
    len(df_copy['year'].copy()),
    len(miles),
    len(exterior_color),
    len(interior_color),
    len(accidents),
    len(owners),
    len(cleaned_price)
)
```

```python
# so, finally all are of same length
```

```python
# so, now w'll make a clean dataframe
```

```python
year = df_copy['year']
```

```python
year
```

```python
df_clean = pd.DataFrame({
                         "make": make,
                         "model": model,
                         "year": year,
                         "miles": miles,
                         "exterior-color": exterior_color,
                         "interior-color": interior_color,
                         "accidents-reported": accidents,
                          "num-of-owners": owners,
                          "price": cleaned_price
                         }
                         )
df_clean
```

```python
df_clean.info()
```

```python
# look at all the unique values in the dataset:
```

```python
print(
        'make:{}'.format(df_clean['make'].unique()),
       'model:{}'.format(df_clean['model'].unique()),
        'year:{}'.format(df_clean['year'].unique()),
        'miles:{}'.format(df_clean['miles'].aggregate(func=['mean'])),
        'exterior-color:{}'.format(df_clean['exterior-color'].unique()),
        'interior-color:{}'.format(df_clean['interior-color'].unique()),
        'accidents-reported:{}'.format(df_clean['accidents-reported'].unique()),
        'num-of-owners:{}'.format(df_clean['num-of-owners'].unique()),
        'price:{}'.format(df_clean['price'].aggregate(func=['mean'])),
    sep = "\n\n"
)
```

```python
# now our dataset is clean and ready for data analysis
```

```python
# simple seaborn box plot by using the price as datapoints
```

```python
box = sns.boxplot(
    x=df_clean['price'].sort_values()
)
plt.show()
print(df_clean['price'].sort_values().describe())
```

```python
# 1st quartile = Q1 = 17991
# 3rd quartile = Q3 = 30999
# using these we remove outliers
```

```python
# grouby accidents-reported and aggregate price using mean
```

```python
# this graph is before removing outliers
sns.barplot(
    x='accidents-reported',
    y='price',
    data=df_clean.groupby('accidents-reported')['price'].mean().to_frame().reset_index(),
    palette = sns.color_palette('magma',n_colors=6)
)

plt.title('mean price by accidents reported')
```

```python
# now, w'll remove outliers
```

```python
def return_lower_upper_tails(q1,q3):
    IQR = q3-q1
    lower = q1 - (1.5*IQR)
    upper = q3 + (1.5*IQR)
    return lower,upper

```

```python
# now make copy of our dataframe
df_removed_outliers_by_price = df_clean.copy()
```

```python
lower_tail,upper_tail = return_lower_upper_tails(q1=17991,q3=30999.25)
```

```python
lower_tail
```

```python
upper_tail
```

```python
index_q1 = df_removed_outliers_by_price[(df_removed_outliers_by_price['price']<lower_tail)].index
index_q3 = df_removed_outliers_by_price[(df_removed_outliers_by_price['price']>upper_tail)].index
```

```python
index_q1
```

```python
index_q3
```

```python
df_removed_outliers_by_price.drop(index=index_q1,inplace=True)
df_removed_outliers_by_price.drop(index=index_q3,inplace=True)
```

```python
df_removed_outliers_by_price
```

```python
box = sns.boxplot(
    x=df_removed_outliers_by_price['price'].sort_values()
)
plt.show()
print(df_removed_outliers_by_price['price'].sort_values().describe())
```

```python
# Above we can see that we have successfully removed the outliers 
# (I think some are still there because seaborn calculates them differently but we have removed the extreme ones so we're safe)!
```

```python
# this graph is before removing outliers
sns.barplot(
    x='accidents-reported',
    y='price',
    data=df_removed_outliers_by_price.groupby('accidents-reported')['price'].mean().to_frame().reset_index(),
    palette = sns.color_palette('magma',n_colors=6)
)

plt.title('mean price by accidents reported')
```
